```python
import pandas as pd
import shap
import lime
import lime.lime_tabular
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

fraud_data = pd.read_csv('C:/Users/nejat/AIM Projects/week8 data/Fraud_Data_cleaned.csv')

X_fraud = fraud_data.drop(columns=['class'])
y_fraud = fraud_data['class']
```

```python
rf_model_fraud = RandomForestClassifier(n_estimators=100)
rf_model_fraud.fit(X_fraud, y_fraud)
```

```python
explainer_fraud = shap.TreeExplainer(rf_model_fraud)
shap_values_fraud = explainer_fraud.shap_values(X_fraud)

shap.summary_plot(shap_values_fraud[1], X_fraud)  # 1 for fraudulent predictions
```
